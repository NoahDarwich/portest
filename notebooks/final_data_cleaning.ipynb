```python
import pandas as pd
```

```python
iraq_df = pd.read_csv('../pro-test/data/iraq_1_drop_non_impact_params.csv', index_col=0)
leb_df = pd.read_csv('../pro-test/data/Leb_1_drop_non_impact_params.csv', index_col=0)
egypt_df = pd.read_csv('../pro-test/data/egypt_1_drop_non_impact_params.csv', index_col=0)
```

```python
egypt_df = pd.read_csv('../pro-test/data/egypt_1_drop_non_impact_params.csv', index_col=0)
```

```python
iraq_df['country'] = 'Iraq'
leb_df['country'] = 'Lebanon'
```

```python
egypt_df['country'] = 'Egypt'
```

```python
mutual_cols = []
special_cols_iraq = []
for column in iraq_df.columns:
    if column in leb_df.columns or column in egypt_df.columns:
        mutual_cols.append(column)
    else:
        special_cols_iraq.append(column)
```

```python
cleaned_iraq = iraq_df.drop(columns=special_cols_iraq)
```

```python
cleaned_iraq = cleaned_iraq.drop(columns=['nameofevent','endate','gpslatstart', 'gpslongstart', 'locationtypestart','slogans','organizer','orgtypesecond','youth/students', 'laborers/workers', 'peasants/farmers',
       'party members', 'activists/movts', 'gov workers', 'business',
       'residents', 'soldiers', 'police', 'participantother', 'sector',
       'industryfirst', 'industrysecond','campaign'])
```

```python
special_cols_leb = []
for column in leb_df.columns:
    if column in iraq_df.columns or column in egypt_df.columns:
        mutual_cols.append(column)
    else:
        special_cols_leb.append(column)
```

```python
special_cols_leb
```

```python
cleaned_leb = leb_df.drop(columns=special_cols_leb)
```

```python
cleaned_leb = cleaned_leb.drop(columns=['nameofevent','endate','gpslatstart', 'gpslongstart', 'locationtypestart','slogans','organizer','orgtypesecond','sector', 'participantother', 'sector', 'industryfirst', 'industrysecond','campaign'])
```

```python
special_cols_egypt = []
for column in egypt_df.columns:
    if column in iraq_df.columns or column in leb_df.columns:
        mutual_cols.append(column)
    else:
        special_cols_egypt.append(column)
```
